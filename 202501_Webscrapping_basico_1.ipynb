```python
# Importamos las bibliotecas
import requests
from bs4 import BeautifulSoup

# URL de la página a scrapear
url = "https://quotes.toscrape.com/"

# Hacemos la solicitud HTTP para obtener el HTML de la página
response = requests.get(url)

# Verificamos que la solicitud fue exitosa
if response.status_code == 200:
    print("Conexión exitosa.")
    # Creamos un objeto BeautifulSoup para analizar el HTML
    soup = BeautifulSoup(response.content, "html.parser")

    # Buscamos todas las citas (etiquetas <span> con clase "text")
    quotes = soup.find_all("span", class_="text")
    
    # Iteramos y mostramos cada cita
    for quote in quotes:
        print("Cita:", quote.text)
else:
    print("Error al conectar con la página.")
```

```python
# Extraer citas, autores y etiquetas de cada bloque
import requests
from bs4 import BeautifulSoup

# URL de la página
url = "https://quotes.toscrape.com/"

# Realizamos la solicitud HTTP
response = requests.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.content, "html.parser")
    
    # Encontramos todos los bloques de citas
    quotes = soup.find_all("div", class_="quote")

    # Iteramos sobre cada cita para extraer información
    for quote in quotes:
        text = quote.find("span", class_="text").text  # Texto de la cita
        author = quote.find("small", class_="author").text  # Autor
        tags = [tag.text for tag in quote.find_all("a", class_="tag")]  # Etiquetas
        
        # Imprimimos la información extraída
        print("Cita:", text)
        print("Autor:", author)
        print("Etiquetas:", ", ".join(tags))
        print("-" * 50)  # Separador
else:
    print("Error al conectar con la página.")
```
