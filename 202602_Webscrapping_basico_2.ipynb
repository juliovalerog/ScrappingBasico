```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL de la página
url = "https://books.toscrape.com/catalogue/page-1.html"

# Hacemos la solicitud
response = requests.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.content, "html.parser")
    books = soup.find_all("article", class_="product_pod")  # Bloques de libros

    # Almacenamos la información en una lista
    book_data = []
    for book in books:
        title = book.find("h3").find("a")["title"]  # Título del libro
        price = book.find("p", class_="price_color").text  # Precio
        availability = book.find("p", class_="instock availability").text.strip()  # Disponibilidad
        book_data.append({"Título": title, "Precio": price, "Disponibilidad": availability}) # Añadimos a la lista

    # Convertimos a DataFrame
    df = pd.DataFrame(book_data)
    print(df)
else:
    print("Error al conectar con la página.")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL de la página
base_url = "https://books.toscrape.com/catalogue/page-{}.html" # URL base

all_books = []  # Lista para almacenar los datos de todos los libros

for page in range (1, 4):  # Iteramos sobre las primeras 3 páginas
    response = requests.get(base_url.format(page))  # URL de la página actual
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, "html.parser")
        books = soup.find_all("article", class_="product_pod")

        # Almacenamos la información en una lista
        for book in books:
            title = book.find("h3").find("a")["title"]  # Título del libro
            price = book.find("p", class_="price_color").text  # Precio
            availability = book.find("p", class_="instock availability").text.strip()  # Disponibilidad
            all_books.append({"Título": title, "Precio": price, "Disponibilidad": availability}) # Añadimos a la lista

    else:
        print("Error al conectar con la página.")  # Mensaje de error

# Convertimos a DataFrame
df = pd.DataFrame(all_books)
print(df)

df.to_csv("books.csv", index=False) # Guardamos los datos en un archivo CSV
print("Datos guardados en 'books.csv'.")
```
